# Crop and Fertilizer Recommendation System

This project involves analyzing the datasets `Crop_Recommendation.csv` and `Fertilizer_Prediction.csv`. Below are the steps for loading, exploring, and visualizing the data.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the datasets

```python
crop_data = pd.read_csv('Crop_Recommendation.csv')
fertilizer_data = pd.read_csv('Fertilizer_Prediction.csv')
```

## Displaying the first and last few rows of the datasets

```python
print("Crop Dataset - Head:")
print(crop_data.head())
print("\nCrop Dataset - Tail:")
print(crop_data.tail())

print("\nFertilizer Dataset - Head:")
print(fertilizer_data.head())
print("\nFertilizer Dataset - Tail:")
print(fertilizer_data.tail())
```

## Checking the shape of the datasets

```python
print("\nCrop Dataset Shape:", crop_data.shape)
print("Fertilizer Dataset Shape:", fertilizer_data.shape)
```

## Displaying dataset information

```python
print("\nCrop Dataset Info:")
crop_data.info()
print("\nFertilizer Dataset Info:")
fertilizer_data.info()
```

## Checking for missing values

```python
print("\nMissing Values in Crop Dataset:")
print(crop_data.isnull().sum())
print("\nMissing Values in Fertilizer Dataset:")
print(fertilizer_data.isnull().sum())
```

## Checking for duplicate values

```python
print("\nDuplicate Values in Crop Dataset:", crop_data.duplicated().sum())
print("Duplicate Values in Fertilizer Dataset:", fertilizer_data.duplicated().sum())
```

## Statistical summary of the datasets

```python
print("\nCrop Dataset Statistics:")
print(crop_data.describe())
print("\nFertilizer Dataset Statistics:")
print(fertilizer_data.describe())
```

## Displaying the column names

```python
print("\nCrop Dataset Columns:", crop_data.columns)
print("Fertilizer Dataset Columns:", fertilizer_data.columns)
```

## Checking the target feature distribution

```python
if 'label' in crop_data.columns:
    print("\nCrop Target Feature Distribution:")
    print(crop_data['label'].value_counts())
    sns.countplot(x='label', data=crop_data)
    plt.title("Crop Target Feature Distribution")
    plt.show()

if 'label' in fertilizer_data.columns:
    print("\nFertilizer Target Feature Distribution:")
    print(fertilizer_data['label'].value_counts())
    sns.countplot(x='label', data=fertilizer_data)
    plt.title("Fertilizer Target Feature Distribution")
    plt.show()
```

## Data visualization - Correlation heatmap for crop dataset

```python
# Selecting only numeric columns for correlation
numeric_columns = crop_data.select_dtypes(include=['number'])

# Checking if there are numeric columns
if not numeric_columns.empty:
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_columns.corr(), annot=True, cmap="coolwarm", fmt='.2f')
    plt.title("Correlation Heatmap - Crop Dataset")
    plt.show()
else:
    print("No numeric columns available in the dataset for correlation analysis.")
```

## Pairplot for numerical columns in crop dataset

```python
sns.pairplot(crop_data, diag_kind='kde')
plt.title("Pairplot - Crop Dataset")
plt.show()
```

## Data visualization - Correlation heatmap for fertilizer dataset

```python
# Selecting only numeric columns for correlation
numeric_columns_fertilizer = fertilizer_data.select_dtypes(include=['number'])

# Checking if there are numeric columns
if not numeric_columns_fertilizer.empty:
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_columns_fertilizer.corr(), annot=True, cmap="coolwarm", fmt='.2f')
    plt.title("Correlation Heatmap - Fertilizer Dataset")
    plt.show()
else:
    print("No numeric columns available in the dataset for correlation analysis.")
```

## Feature distributions (Histograms for numerical columns)

```python
for column in crop_data.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(crop_data[column], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {column} - Crop Dataset")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()

for column in fertilizer_data.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(fertilizer_data[column], kde=True, bins=30, color='green')
    plt.title(f"Distribution of {column} - Fertilizer Dataset")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
